```python
# Import the random module.   6.1.4
import random
```

```python
# This function returns only a floating-point decimal number between 0 and 1.0.
random.random()
```

```python
# short for "random integer." 
# generates a single integer 
random.randint(-90, 90)
```

```python
random_number = random.randint(-90, 89) + random.random()
random_number
```

```python
x = 1
latitudes = []
while x < 11:
    random_lat = random.randint(-90, 89) + random.random()
    latitudes.append(random_lat)
    x += 1
latitudes
```

```python
# Returns a whole integer inside () where the step is the difference between each number in the sequence.
random.randrange(-90, 90, step=1)
```

```python
# Returns a floating-point decimal number between the interval, -90 and 90.
random.uniform(-90, 90)
```

```python
# Import the NumPy module.
import numpy as np
```

```python
# Option one. Adding the 0s after the . is not necerasry
np.random.uniform(-90.000, 90.000)
```

```python
#Option two. the parenthetical parameters contain a lower boundary (low value) and an upper boundary (high value) that are floating-point decimal numbers.
np.random.uniform(low=-90, high=90)
```

```python
# To generate more than one floating-point decimal number
np.random.uniform(-90. , 90., size=50)
```

```python
# Import timeit.
import timeit
```

```python
# Checking how long it takes to run the code option 1
%timeit np.random.uniform(-90.000, 90.000, size=1500)
```

```python
# Checking how long it takes to run the code option 2
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes
# Call the function with 1500.
%timeit latitudes(1500)
```
